```python
# Importing the Pandas package and reading the CSV file and storing in a dataframe

import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
df
```

## 1. Delete unnamed columns  

```python
# Deleting the unnamed column from the df dataframe. case keyword is false to make independent of upper/lower case of the unammed column 
# It will check the column names which is having the 'unnamed' and delete the column
df=df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1)
df
```

## 2. Show the distribution of male and female 

```python
# Grouping the data(df) using the Year and Gender columns and calculating the count of male and female in respective year
df_genderDist=pd.DataFrame(df.groupby(['Year','Gender']).sum()['Count'])
df_genderDist
```

```python
import matplotlib.pyplot as plt                                       
fig, ax = plt.subplots(figsize=(15,5))                               # Defininga one subplot with figure size (15,5)
df.groupby(['Year','Gender']).sum()['Count'].unstack().plot(ax=ax)   # Grouping the data based on Year and Gender and ploting it 
ax.set_xlabel('State')                                               # Defining the X - axis label
ax.set_ylabel('Number of Male/Female')                               # Defining the Y - axis label
```

## 3. Show the top 5 most preferred names 

```python
# Sotring the Dataframe(df) based up on Count and ararnging in Descending order and taking top 
dfSort=pd.DataFrame(df.groupby(['Name']).sum()['Count'])    # Grouping the data based up on Names  
dfSort=dfSort.sort_values(by="Count",ascending=False)       # Sorting the data based on count in descending order
dfSort.head()                                               # Display the top 5 rows
```

## 4. What is the median name occurence in the dataset 

```python
# Calculating the Names median Occurances using the median function
df_median=df
df_median[df_median.Count == df_median.Count.median()]
```

```python

# Calculating the median value based on grouping all the names arranging the count in descending and calculating the median on overall dataset

dfSorttotal=dfSort                            # Copying dfsort into new dataframe 
dfSorttotal["Names"]=dfSorttotal.index        # Creating a new column using the index values 

columnsTitles=["Names","Count"]                              
dfSorttotal=dfSorttotal.reindex(columns=columnsTitles)     # Swaping the column(Names and Count) positions 
dfSorttotal=dfSorttotal.reset_index(drop=True)             # Reseting  the index
totalRows = dfSorttotal.shape[0]                           # Checking the total number of rows
print("Total No of Rows", totalRows)
if((totalRows % 2) == 0):                                  #  if the total number of rows are even
    firstposition=(totalRows/2)                            # calculate the first position
    secposition=(totalRows / 2)+1                          # calculate the Second position
    firstVal=dfSorttotal["Names"][firstposition-1]         # calculate the first position Value
    SecVal=dfSorttotal["Names"][secposition-1]             # calculate the second position Value
    print("Median cant be calculated on string (Names), as the number of rows are even")
    print("Name value at index (N/2)-1 is ",firstVal)       
    print("Name value at index (N/2) is ",SecVal)
else:                                                      # if the total number of rows are even
    positionVal=int(totalRows / 2)+1                       # calculatng the middle position
    Median=dfSorttotal["Names"][positionVal-1]             # finding the median name from the data frame
    print("Medain index value is ",positionVal-1)
    print("Medain Name occurance value is ",Median)
```
